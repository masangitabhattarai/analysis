Detecting and Removing Outliers

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

df = remove_outliers(df, 'adr')
df = remove_outliers(df, 'lead_time')
```

```python
# Check the shape before removing outliers
before_shape = df.shape

# Remove outliers
df = remove_outliers(df, 'adr')
df = remove_outliers(df, 'lead_time')

# Check the shape after removing outliers
after_shape = df.shape

print("Shape before removing outliers:", before_shape)
print("Shape after removing outliers:", after_shape)
print("Number of rows removed:", before_shape[0] - after_shape[0])
```

```python
# Boxplot before removing outliers
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['adr'])
plt.title("ADR Distribution Before Removing Outliers")
plt.subplot(1, 2, 2)
sns.boxplot(x=df['lead_time'])
plt.title("Lead Time Distribution Before Removing Outliers")
plt.show()
```

```python
# Function to check for outliers
def check_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] < lower) | (df[column] > upper)]
# Check for remaining outliers in 'adr' and 'lead_time'
remaining_outliers_adr = check_outliers(df, 'adr')
remaining_outliers_lead_time = check_outliers(df, 'lead_time')
print("Remaining outliers in ADR:", remaining_outliers_adr.shape[0])
print("Remaining outliers in Lead Time:", remaining_outliers_lead_time.shape[0])
```

```python
# Remove outliers again for ADR
df = remove_outliers(df, 'adr')
# Check for remaining outliers again
remaining_outliers_adr = check_outliers(df, 'adr')
print("Remaining outliers in ADR after second removal:", remaining_outliers_adr.shape[0])
```
